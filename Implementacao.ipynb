```python
def funcao(x):
  return -(183000*(1/(x+1)**30 - 1))/x - 21600
```

```python
def funcao_diferenciada(x):
  return 5490000/(x*(x+1)**31)+(183000*(1/(x+1)**30 - 1))/x**2
```

```python
def calcula_erro(a, b):
  return abs(b-a)
```

```python
# Bissecção
def bisseccao(a, b, tolerancia, numero_max_interacoes):
  erro = calcula_erro(a, b)
  numero_iteracoes_realizadas = 0

  if (funcao(a) * funcao(b) > 0):
    print("Não há garantia de convergência.")
    return
  
  while ((erro > tolerancia) and (numero_iteracoes_realizadas <= numero_max_interacoes)):
    ponto_medio = (a + b)/2

    if (funcao(a) * funcao(ponto_medio) < 0):
      b = ponto_medio
      erro = calcula_erro(a, b)

    elif (funcao(b) * funcao(ponto_medio) < 0):
      a = ponto_medio
      erro = calcula_erro(a, b)

    numero_iteracoes_realizadas+=1

  print(f"Erro: {erro}")
  print(f"Ponto a: {a}")
  print(f"Ponto b: {b}")
  print(f"Numero iteracoes realizadas: {numero_iteracoes_realizadas}")
```

```python
'''
Dados da célula
a = 1
b = 100
tolerância = 10⁻⁵
Número máximo de iterações: 50
'''

import time
inicio = time.time()
bisseccao(1, 100, 10**(-5), 50)
fim = time.time()
print(f"Tempo de execucao: {fim - inicio}")
```

```python
# Newton
def newton(a, tolerancia, numero_max_interacoes):
  numero_iteracoes_realizadas = 0
  for i in range(0, numero_max_interacoes):
    numero_iteracoes_realizadas+=1
    if (i == 0):
      valor = a - (funcao(a)/funcao_diferenciada(a))
      valor_antigo = valor
      a = valor
    else:
      valor = a - (funcao(a)/funcao_diferenciada(a))
      a = valor
      erro = calcula_erro(a, valor_antigo)
      if(erro < tolerancia or funcao(a) < tolerancia):
        break

  print(f"Erro: {erro}")
  print(f"Ponto a: {a}")
  print(f"Numero iteracoes realizadas: {numero_iteracoes_realizadas}")  
```

```python
# Secantes
def secante(a, b, tolerancia, numero_max_interacoes):
  numero_iteracoes_realizadas = 0
  for i in range(0, numero_max_interacoes):
    funcao_a = funcao(a)
    funcao_b = funcao(b)

    erro = calcula_erro(funcao_a, funcao_b)
    if (erro <= tolerancia):
      print(f"Erro: {erro}")
      print(f"Ponto a: {a}")
      print(f"Ponto b: {b}")
      print(f"Numero iteracoes realizadas: {numero_iteracoes_realizadas}") 
      return

    ponto = a - (funcao_a / ((funcao_a - funcao_b)/(a-b)))

    a = b
    b = ponto

    numero_iteracoes_realizadas+=1

  print(f"Erro: {erro}")
  print(f"Ponto a: {a}")
  print(f"Ponto b: {b}")
  print(f"Numero iteracoes realizadas: {numero_iteracoes_realizadas}") 
```

Neste exercício, implemente o método de Gauss-Seidel para resolução de sistemas lineares de forma a verificar se que há garantia de convergência em relação aos critérios de linhas e colunas (Bônus: verifique também para Sassenfeld). Ou seja, a implementação deve conter uma função que avalie se a matriz de coeficientes satisfaz ao menos algum critério. O método deve ser aplicado aos sistemas como eles estão. Mude
as aproximações iniciais e analise a convergência de acordo com o fato de haver ou não garantia de convergência
para o sistema. É obrigatório que a implementação apresente 2 critérios de parada, em que um deles seja o número máximo de iterações.



```python
import numpy as np
from scipy.linalg import solve

# Critério de linhas
def convergencia_linha(A):
  for i in range(0, A.shape[0]):
    soma = 0
    diagonal = 0
    for j in range(0, A.shape[1]):
      if (i!=j):
        soma += abs(A[i][j])
      else:
        diagonal = abs(A[i][j])
    if (soma > diagonal):
      print("Não satisfaz o critério de linhas")
      return False
  return True

#Critério de colunas
def convergencia_coluna(A):
  for i in range(0, A.shape[1]):
    soma = 0
    diagonal = 0
    for j in range(0, A.shape[0]):
      if (i!=j):
        soma += abs(A[i][j])
      else:
        diagonal = abs(A[i][j])
    if (soma > diagonal):
      print("Não satisfaz o critério de colunas")
      return False
  return True

#Gauss Seidel
def gauss_seidel(A, B, x, numero_max_interacoes):
  if(convergencia_linha(A)):
    print("Satisfaz o critério de linhas")

  if (convergencia_coluna(A)):
    print("Satisfaz o critério de colunas")
  
  L = np.tril(A)
  U = A - L
  for i in range(numero_max_interacoes):
    x = np.dot(np.linalg.inv(L), B-np.dot(U,x))
  return x
```
